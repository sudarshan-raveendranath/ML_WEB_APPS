```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('heart_disease_data.csv')
```

```python
X = data.drop(columns='target', axis=1)
Y = data['target']
```

```python
print(X)
print(Y)
```

```python
print(X.shape, Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
training_prediction = model.predict(X_train)
training_accuracy = accuracy_score(training_prediction, Y_train)
print('Accuracy on training data : ', training_accuracy)
```

```python
testing_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(testing_prediction, Y_test)
print('Accuracy on testing data : ', testing_accuracy)
```

```python
input_data = np.array([[41,0,1,130,204,0,0,172,0,1.4,2,0,2]])
input_data.reshape(1, -1)
prediction = model.predict(input_data)
print(prediction)
```
